# Объем данных на auto.ru

```python
# Функция возвращает кол-во страниц с объявлениями для каждой модели
def amount_of_pages(row):
    try:
        brand = row.brand
        model = row.model_name
        url = 'https://auto.ru/cars/'+brand+'/'+model+'/used/'
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        page = BeautifulSoup(response.text, 'html.parser')
        container = page.find('div', class_="ListingPagination-module__container")
        txt = container.find('span', class_="ControlGroup ControlGroup_responsive_no ControlGroup_size_s ListingPagination-module__pages").text
        pattern = re.compile(r'\d+')
        time.sleep(1)
        numbers_list = pattern.findall(txt)
        if len(numbers_list)>1:
            return numbers_list[1]
        else:
            if len(numbers_list[0])<10:
                return numbers_list[0][-1]
            else:
                return numbers_list[0][-2:]
    except:
        time.sleep(1)
        return '1'
```

```python
# Применение функции
brand_model['amount_of_pages'] = brand_model.progress_apply(amount_of_pages, axis=1)
```

```python
# Перевод в числовой формат
brand_model['amount_of_pages'] = brand_model.amount_of_pages.astype(int)
```

```python
brand_model.sample(3)
```

```python
# Функция расчитывает кол-во выбросов по столбцам и их долю
def outbursts(df, col):
    Q25 = df[col].quantile(0.25)
    Q75 = df[col].quantile(0.75)
    IQR = Q75 - Q25
    lowest = Q25 - 1.5*IQR
    highest = Q75 + 1.5*IQR
    amount = df[(df[col] < lowest)|(df[col] > highest)].count()[0]
    ratio = amount/df.shape[0]
    return amount, ratio, lowest, highest
```

```python
# Функция выводит основную информацию и график признака
def sum_up_num(df, col):
    print(col.upper())
    print(df[col].describe()[1:])
    print('Кол-во выбросов: {} \t (%%: {})'.format(outbursts(df, col)[0], 
                                                   np.round(outbursts(df, col)[1]*100,2)))
    print()
    fig, axs = plt.subplots(figsize=(15, 5), ncols=2)
    sns.distplot(df[col], kde=True, ax=axs[0])
    sns.boxplot(data = df,  x = col, palette = 'viridis', ax=axs[1])
    plt.show()
```

# Минимальная выборка

Выгружаются максимум 38 объявлений для каждой модели.

Минимальная выборка необходима для начала работы с данными и моделирования. Позже ее можно дополнить.

```python
# Функция возвращает ссылки на страницы конкретных авто с первой страницы моделей
def links_lite(row):
    brand = row.brand
    model = row.model_name
    url = 'https://auto.ru/cars/'+brand+'/'+model+'/used/'
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    page = BeautifulSoup(response.text, 'lxml')
    try:
        tags = page.find_all('a', "Link ListingItemTitle-module__link")
        links = []
        for i in range(len(tags)):
            links.append(tags[i]['href'])
        time.sleep(1)
        return links
    except:
        time.sleep(1)
        return None
```

```python
# Применение функции
brand_model['links_lite'] = brand_model.progress_apply(links_lite, axis=1)
```

```python
# Общее кол-во ссылок
total_links = brand_model.links_lite.sum()
len(total_links)
```

```python
# Функция возвращает данные со станицы авто
# Функция выглядит грамоздко. Полезные данные можно было бы достать из одного скрипта
# Тем не менее, данная версия функции возвращает данные в наиболее близком формате к тестовой выборке с Kaggle
def get_data(link):
    url = link
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    response.encoding = 'utf-8'
    page = BeautifulSoup(response.text, 'lxml')
    
    columns = {}
    general_script = json.loads(page.find('script').string)
    precise_script = json.loads(page.find_all('script')[10].string)
    
    try:
        columns['bodyType'] = general_script['bodyType']
    except:
        columns['bodyType'] = None
    try:
        columns['brand'] = general_script['brand']
    except:
        columns['brand'] = None
    columns['car_url'] = url
    try:
        columns['color'] = page.find('li','CardInfoRow CardInfoRow_color').find_all('span')[1].text
    except:
        columns['color'] = None
    try:
        columns['complectation_dict'] = precise_script['card']['vehicle_info']['complectation']
    except:
        columns['complectation_dict'] = None
    try:
        columns['description'] = general_script['description']
    except:
        columns['description'] = None
    try:
        columns['engineDisplacement'] = general_script['vehicleEngine']['engineDisplacement']
    except:
        columns['engineDisplacement'] = None
    try:
        columns['enginePower'] = general_script['vehicleEngine']['enginePower']
    except:
        columns['enginePower'] = None
    try:
        columns['equipment_dict'] = precise_script['card']['vehicle_info']['equipment']
    except:
        columns['equipment_dict'] = None
    try:
        columns['fuelType'] = general_script['fuelType']
    except:
        columns['fuelType'] = None
    try:
        columns['image'] = general_script['image']
    except:
        columns['image'] = None
    try:
        columns['mileage'] = precise_script['card']['state']['mileage']
    except:
        columns['mileage'] = None
    try:
        columns['modelDate'] = general_script['modelDate']
    except:
        columns['modelDate'] = None
    try:
        columns['model_info'] = precise_script['card']['vehicle_info']['model_info']
    except:
        columns['model_info'] = None
    try:
        columns['model_name'] = precise_script['card']['vehicle_info']['model_info']['code']
    except:
        columns['model_name'] = None
    try:
        columns['name'] = precise_script['card']['vehicle_info']['tech_param']['human_name']
    except:
        columns['name'] = None
    try:
        columns['numberOfDoors'] = general_script['numberOfDoors']
    except:
        columns['numberOfDoors'] = None
    columns['parsing_unixtime'] = int(time.time())
    try:
        columns['priceCurrency'] = general_script['offers']['priceCurrency']
    except:
        columns['priceCurrency'] = None
    try:
        columns['productionDate'] = general_script['productionDate']
    except:
        columns['productionDate'] = None
    try:
        columns['sell_id'] = precise_script['card']['id']
    except:
        columns['sell_id'] = None
    try:
        columns['super_gen'] = precise_script['card']['vehicle_info']['super_gen']
    except:
        columns['super_gen'] = None
    try:
        columns['vehicleConfiguration'] = general_script['vehicleConfiguration']
    except:
        columns['vehicleConfiguration'] = None
    try:
        columns['vehicleTransmission'] = general_script['vehicleTransmission']
    except:
        columns['vehicleTransmission'] = None
    try:
        columns['vendor'] = precise_script['card']['vehicle_info']['vendor']
    except:
        columns['vendor'] = None
    try:
        columns['Владельцы'] = (page.find('li','CardInfoRow CardInfoRow_ownersCount').find_all('span')[1].text).replace("\xa0", " ")
    except:
        columns['Владельцы'] = None
    try:
        columns['Владение'] = (page.find('li','CardInfoRow CardInfoRow_owningTime').find_all('span')[1].text).replace("\xa0", " ")
    except:
        columns['Владение'] = None
    try:
        columns['ПТС'] = page.find('li','CardInfoRow CardInfoRow_pts').find_all('span')[1].text
    except:
        columns['ПТС'] = None
    try:
        columns['Привод'] = page.find('li','CardInfoRow CardInfoRow_drive').find_all('span')[1].text
    except:
        columns['Привод'] = None
    try:
        columns['Руль'] = page.find('li','CardInfoRow CardInfoRow_wheel').find_all('span')[1].text
    except:
        columns['Руль'] = None
    try:
        columns['Состояние'] = page.find('li','CardInfoRow CardInfoRow_state').find_all('span')[1].text
    except:
        columns['Состояние'] = None
    try:
        columns['Таможня'] = page.find('li','CardInfoRow CardInfoRow_customs').find_all('span')[1].text
    except:
        columns['Таможня'] = None
    try:
        columns['price_info'] = precise_script['card']['price_info']
    except:
        columns['price_info'] = None
    
    time.sleep(1.5)
    
    return pd.Series(columns)
```

```python
# Создание DataFrame для полученных данных
train_lite = pd.DataFrame(columns=['bodyType', 'brand', 'car_url', 'color', 'complectation_dict',
       'description', 'engineDisplacement', 'enginePower', 'equipment_dict', 'fuelType', 'image',
       'mileage', 'modelDate', 'model_info', 'model_name', 'name',
       'numberOfDoors', 'parsing_unixtime', 'priceCurrency', 'productionDate',
       'sell_id', 'super_gen', 'vehicleConfiguration', 'vehicleTransmission',
       'vendor', 'Владельцы', 'Владение', 'ПТС', 'Привод', 'Руль', 'Состояние',
       'Таможня', 'price_info'])

# Построчная запись полученных данных
for link in tqdm(total_links[train_lite.shape[0]:]):
    train_lite = train_lite.append(get_data(link), ignore_index=True)
```

# Дополнение выборки

```python
# Функция возвращает ссылки на страницы конкретных авто с последующих страниц моделей
# Максимум 20 страниц
def add_links(row):
    if row.amount_of_links<2:
        return None
    else:
        if row.amount_of_links>20:
            limit = 21
        else:
            limit = row.amount_of_links+1
        links = []
        for i in range(2,limit):
            brand = row.brand
            model = row.model_name
            url = 'https://auto.ru/cars/'+brand+'/'+model+'/used/?page='+str(i)
            response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
            page = BeautifulSoup(response.text, 'lxml')
            try:
                tags = page.find_all('a', "Link ListingItemTitle-module__link")
                for i in range(len(tags)):
                    links.append(tags[i]['href'])
                time.sleep(1)
            except:
                links.append(None)
                time.sleep(1)
        return links
```

```python
# Применение функции
# Данный способ позволяет сохранять выгруженные ссылки после каждой итерации
additional_links = pd.DataFrame(columns=['additional_links'])
for i in tqdm(brand_model.index):
    additional_links = additional_links.append(pd.Series({'additional_links': add_links(brand_model.loc[i])}), 
                            ignore_index=True)
```

```python
# Добавление в основной DataFrame со ссылками
brand_model['additional_links'] = additional_links
brand_model.sample(3)
```

```python
# Проверка, что ссылки не выгружены только для моделей с одной страницей
brand_model[(brand_model.additional_links.isna()==True)|(len(brand_model.additional_links)==0)].amount_of_pages.unique()
```

```python
# Кол-во дополнительных ссылок
add_total_links = brand_model.additional_links.dropna().sum()
len(add_total_links)
```
