```python
# Загрузка библиотек
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import silhouette_score, mean_absolute_error
```

```python
# Фиксация random seed для воспроизводимости экспериментов
RS = 42
```

```python
# Загрузка данных
model_selection = pd.read_csv('model_selection.csv')
model_selection.sample(3)
```

```python
# Разбиение выборки
X = model_selection.drop('Rating', axis = 1)
y = model_selection.Rating
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = RS)
```

## RandomForest

```python
results = []
count = 1
for ne in [25, 50, 75, 100, 500, 1000]:
    RF = RandomForestRegressor(random_state=RS, n_estimators=ne)
    RF.fit(X_train, y_train)
    results.append(mean_absolute_error(y_test, RF.predict(X_test)))
    print(('Completed iter: {}'.format(count)), end="\r")
    count+=1

plt.figure(figsize=(15, 5))
ax = sns.lineplot(x=[25, 50, 75, 100, 500, 1000], y=results, marker='o')
ax.set(ylabel='MAE', xlabel='n_estimators')
plt.show()
```

```python
results = []
count = 1
for md in [3, 5, 10, 25, 100, None]:
    RF = RandomForestRegressor(random_state=RS, n_estimators=1000, max_depth=md)
    RF.fit(X_train, y_train)
    results.append(mean_absolute_error(y_test, RF.predict(X_test)))
    print(('Completed iter: {}'.format(count)), end="\r")
    count+=1
plt.figure(figsize=(15, 5))
ax = sns.lineplot(x=[3, 5, 10, 25, 100, 150], y=results, marker='o')
ax.set(ylabel='MAE', xlabel='max_depth')
plt.show()
```

## GradientBoosting

```python
results = []
count = 1
for ls in ['ls', 'lad']:
    GBR = GradientBoostingRegressor(random_state=RS, loss=ls)
    GBR.fit(X_train, y_train)
    results.append(mean_absolute_error(y_test, GBR.predict(X_test)))
    print(('Completed iter: {}'.format(count)), end="\r")
    count+=1

plt.figure(figsize=(15, 5))
ax = sns.lineplot(x=['ls', 'lad'], y=results, marker='o')
ax.set(ylabel='MAE', xlabel='loss')
plt.show()
```
